```python
"""SQLAlchemy is a popular SQL toolkit and Object Relational Mapper
SQLAlchemy is famous for its object-relational mapper (ORM), using which classes can be mapped to the database
Thereby allowing the object model and database schema to develop in a cleanly decoupled way from the beginning."""
```

```python
#Create an sqlalchemy engine using a sample from the data set

import sqlalchemy
from sqlalchemy import MetaData,Table,Column,String,Integer,create_engine
engine=create_engine('sqlite:///C:\\Users\\Chidi\\sqlite\\sqlitee\'suvey.db', echo=True)
connection=engine.connect()
meta=MetaData()
sensextable=Table(
 'sensextable',meta,
    Column('age',Integer),
    Column('occupation',String),
    Column('relationship',String),
    Column('workclass',String),
    
    )
    
meta.create_all(engine)   
```

```python
#Populating rows of the table thru insert() query

connection=engine.connect()
connection.execute(sensextable.insert(), [
   {'age':30,'occupation' : 'handlers-cleaners','relationship':'husband' , 'workclass':'Private'},
   {'age':43,'occupation':'Adm-clerical' , 'relationship' :'wife', 'workclass':'stategov'},
   {'age':33,'occupation':'Exec-Manager','relationship' : 'not-in-family','workclass':'Private'},
   {'age':50,'occupation':'cleaner','relationship' : 'son','workclass':'self-employ'},
])
```

```python
#fetching all the rows of the table object
s=sensextable.select()
connection=engine.connect()
result=connection.execute(s)
for row in result:
    print(row)
```

```python
# updating the values of  the relationship column in the table, where the age is 39 
    
connection = engine.connect()
stmt=sensextable.update().where(sensextable.c.age==39).values(relationship='sister-in-law')
connection.execute(stmt)
s = sensextable.select()
connection.execute(s).fetchall()
```

```python
# updating the values of  the workclass column in the table, where the age is 39 
connection = engine.connect()
stmt=sensextable.update().where(sensextable.c.age==39).values(workclass='self-employed')
connection.execute(stmt)
s = sensextable.select()
connection.execute(s).fetchall()
```

```python
# deleting  the row from the table, where the  value of the relationship column is 'not-in-family'
connection = engine.connect()
stmt = sensextable.delete().where(sensextable.c.relationship == 'not-in-family')
connection.execute(stmt)
s = sensextable.select()
connection.execute(s).fetchall()
```

```python
# deleting  the row from the table, where the  value of the age column is 30
connection=engine.connect()
stmt=sensextable.delete().where(sensextable.c.age==30 )
connection.execute(stmt)
s=sensextable.select()
connection.execute(s).fetchall()
```

```python
#filtering the value of the age column,where the age  is > = 42
connection=engine.connect()
stmt=sensextable.select().where(sensextable.c.age >= 42)
connection.execute(stmt).fetchall()
```

```python
#filtering the value of the workclass column,where the category is 'self-employed'
connection=engine.connect()
stmt=sensextable.select().where(sensextable.c.workclass=='self-employed')
connection.execute(stmt).fetchall()
```

```python
#function for finding the current timestamp
from sqlalchemy import *
from sqlalchemy.sql import func
result = connection.execute(select([func.now()]))
print (result.fetchone())
```

```python
# function for finding the average age in the sample sensex table
from sqlalchemy import *
from sqlalchemy.sql import func
result=connection.execute(select([func.avg(sensextable.c.age)]))
print (result.fetchone())
```

```python
#function for converting the case of the workclass column in the sample sensex table
connection=engine.connect()
result=connection.execute(select([func.upper(sensextable.c.workclass)]))
print (result.fetchall())
```
